```python
import seaborn as sns  #another simplier library with inborn datasets
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#box plot: see how things distribute in categoties
plt.figure()
sns.boxplot(x=tips['total_bill'])
```

```python
plt.figure()
sns.boxplot(x = 'day', y = 'total_bill', data = tips)  #define x and y, then tell is where you data is
```

```python
plt.figure()
sns.violinplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, split = True)   #hue divides things into categories
```
